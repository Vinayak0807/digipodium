```python
df.rename(columns={'University Rank':'Ranking','Name of University':'University'},inplace=True)
```

```python
cols_to_drop = ['International Student' , 'No of student per staff' , 'No of student']

df.drop(columns=cols_to_drop , inplace=True)
```

```python
plt.figure(figsize=(10, 6), dpi=300)
plt.bar(df['University'].head(5), df['International Student'].head(5))
plt.xticks(rotation=90)
plt.xlabel('University')
plt.ylabel('Ranking')
plt.title('Top 10 Universities Ranking')
plt.show()
```

```python
top_10 = df.head(10)

plt.figure(figsize=(10, 6), dpi=300)
plt.barh(top_10['University'], top_10['Ranking'])
plt.xlabel('Ranking')
plt.ylabel('University')
plt.title('Top 10 Universities Ranking')
plt.gca().invert_yaxis()  # Reverse the order of universities
plt.show()
```
